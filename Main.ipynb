```python
import cv2  # Importe le module OpenCV pour le traitement d'images
import controller as cnt # Controller est un programme python, pour controler l'arduino //Attention// il faut l'ajouter dans le même répertoire de votre jupyter.
from cvzone.HandTrackingModule import HandDetector  # Importe la classe HandDetector du module "HandTrackingModule" de "cvzone"
```

```python
# Initialise le détecteur de mains avec une précision de détection de 0.8 et une limite d'une main détectée
detector = HandDetector(detectionCon=0.8, maxHands=1)

# Initialise la capture vidéo à partir de la caméra par défaut (index 0)
video = cv2.VideoCapture(0,cv2.CAP_DSHOW)
print("Camera is open")
# Boucle principale
while True:
    # Capture un frame de la vidéo et renvoie un booléen (ret) et le frame
    ret, frame = video.read()
    # Miroir horizontal du frame pour refléter correctement les mouvements de la main
    frame = cv2.flip(frame, 1)

    # Utilise le détecteur de main pour trouver les mains dans le frame
    hands, img = detector.findHands(frame)

    # Vérifie si des mains ont été détectées dans le frame
    if hands:
        # Obtient les coordonnées des points de repère de la première main détectée
        lmList = hands[0]
        # Utilise les coordonnées des points de repère pour détecter si les doigts sont levés ou non
        fingerUp = detector.fingersUp(lmList)
        # Calcule le nombre total de doigts levés
        total = sum(fingerUp)
        # Appelle la fonction "led" du module "controller" avec le nombre total de doigts levés
        cnt.led(total)
        print(fingerUp)
        
            # Affiche le nombre de doigts levés sur le frame en utilisant cv2.putText
        if total == 0:
            cv2.putText(frame, 'Finger count: 0', (20, 460), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 1, cv2.LINE_AA)
        elif total == 1:
            cv2.putText(frame, 'Finger count: 1', (20, 460), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 1, cv2.LINE_AA)
        elif total == 2:
            cv2.putText(frame, 'Finger count: 2', (20, 460), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 1, cv2.LINE_AA)
        elif total == 3:
            cv2.putText(frame, 'Finger count: 3', (20, 460), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 1, cv2.LINE_AA)
        elif total == 4:
            cv2.putText(frame, 'Finger count: 4', (20, 460), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 1, cv2.LINE_AA)
        elif total == 5:
            cv2.putText(frame, 'Finger count: 5', (20, 460), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 1, cv2.LINE_AA)

    # Affiche le frame avec les annotations
    cv2.imshow("frame", frame)
    # Attend l'appui sur une touche pendant 1 milliseconde
    k = cv2.waitKey(1)
    # Taper la touche "k" pour quitter la boucle
    if k == ord("k"):
        break

# Libère la capture vidéo
video.release()
# Ferme toutes les fenêtres OpenCV
cv2.destroyAllWindows()
```
